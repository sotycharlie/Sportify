```python
data['region']=data['region'].str.lower()
country_to_continent = {country.lower(): continent.lower() for country, continent in country_to_continent.items()}
```

```python
data['continent'] = data['region'].map(country_to_continent)
```

```python
group_by_continent = data.groupby(['continent','region'])
```

```python
for(continent, region), group in group_by_continent:
    filename = f"{continent}_{region}.csv".replace(" ", "_")
    group.to_csv(filename, index=False)
```
